# SQL QUERIES 

```python
# -- Write an SQL query to check if all Orders have a valid customer_id present in the Customers table.
q1 = 'SELECT customer_id FROM orders WHERE customer_id NOT IN (SELECT customer_id FROM customers)'
a1=pd.read_sql(q1,conn)
a1
```

```python
# -- Write an SQL query to find unpaid orders (i.e., orders where no successful payment exists).
q2 = 'SELECT * , p.payment_status FROM orders o JOIN payments p ON o.order_id = p.order_id WHERE p.payment_status != "Completed"'
a2=pd.read_sql(q2,conn)
a2
```

```python
# -- How can you verify that each Payment record correctly references an existing order_id?
q3 = 'SELECT order_id  FROM payments  WHERE payment_id NOT IN (SELECT order_id FROM orders)'
a3=pd.read_sql(q3,conn)
a3
```

```python
# -- Retrieve all customer details.
q4 = 'SELECT * FROM customers'
a4=pd.read_sql(q4,conn)
a4
```

```python
# -- List all orders with their total amount and status.
q5 = 'SELECT * FROM orders;'
a5=pd.read_sql(q5,conn)
a5
```

```python
# -- Find all products that belong to the ‘Electronics’ category.
q6 = 'SELECT product_name FROM products WHERE category = "Electronics";'
a6=pd.read_sql(q6,conn)
a6
```

```python
# -- Find the total number of orders placed by each customer (customer_id and count).
q7 = '''SELECT c.customer_id, COUNT(o.customer_id) FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
GROUP BY c.customer_id;'''
a7=pd.read_sql(q7,conn)
a7
```

```python
# -- Retrieve a list of all orders, including customer names and order status.
q8 = """SELECT o.order_id, o.customer_id, c.first_name, c.last_name , o.status
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id;"""
a8=pd.read_sql(q8,conn)
a8
```

```python
# -- Find the total revenue (sum of all orders) generated.
q9 = """SELECT SUM(total_amount) as total_revenue 
FROM orders;"""
a9=pd.read_sql(q9,conn)
a9
```

```python
# -- Identify customers who have placed more than 3 orders.
q10 = """SELECT c.customer_id, c.first_name, c.last_name, COUNT(o.customer_id) as total_orders
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
GROUP BY c.customer_id,c.first_name, c.last_name
HAVING total_orders > 3;"""
a10=pd.read_sql(q10,conn)
a10
```

```python
# -- Retrieve all completed payments and include the associated order details.
q11 = """SELECT *, o.order_date,o.total_amount
FROM payments p
JOIN orders o ON p.order_id = o.order_id
WHERE p.payment_status = "Completed";"""
a11=pd.read_sql(q11,conn)
a11
```

```python
# -- Find the top 5 customers who have spent the most on orders.
q12 = """SELECT c.customer_id, c.first_name, c.last_name, c.email, SUM(o.total_amount) total_spent
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
GROUP BY c.customer_id,c.first_name, c.last_name, c.email
ORDER BY total_spent DESC
LIMIT 5;"""
a12=pd.read_sql(q12,conn)
a12
```

```python
# -- List all orders where the payment was either ‘Pending’ or ‘Failed’.
q13 = """SELECT o.order_id , p.payment_id, p.payment_status
FROM orders o
JOIN payments p ON p.order_id = o.order_id
WHERE p.payment_status = "Pending" OR p.payment_status = "Failed";"""
a13=pd.read_sql(q13,conn)
a13
```

```python
# -- Find the most popular product category based on the number of orders.
q14 = """SELECT p.category, COUNT(o.order_id) AS total_order_count
FROM products  p
JOIN orders o ON p.product_id = o.order_id
GROUP BY p.category
ORDER BY  total_order_count DESC
LIMIT 1;"""
a14=pd.read_sql(q14,conn)
a14
```

```python
# -- Identify orders that were placed but have no successful payment recorded.
q15 = """SELECT o.order_id, o.customer_id, o.total_amount, o.status, p.payment_status
FROM orders AS o
JOIN payments AS p ON p.order_id = o.order_id
WHERE p.payment_status != "Completed";"""
a15=pd.read_sql(q15,conn)
a15
```

```python
# -- Calculate the average order value (AOV) per customer.
q16 = """SELECT c.customer_id, c.first_name, c.last_name, c.email,AVG(o.total_amount) AS average_orders
FROM customers  c
JOIN orders o ON c.customer_id = o.customer_id
GROUP BY c.customer_id,c.first_name, c.last_name, c.email;"""
a16=pd.read_sql(q16,conn)
a16
```

```python
# -- Identify which payment method is most commonly used.
q17 = """SELECT payment_method, COUNT(payment_method) AS count
FROM payments
GROUP BY payment_method
ORDER BY count DESC
LIMIT 1;"""
a17=pd.read_sql(q17,conn)
a17
```

```python
# -- Determine the percentage of orders that were canceled.
q18 = """SELECT (SUM(CASE WHEN status = 'Cancelled' THEN 1 ELSE 0 END) * 100 / COUNT(*)) AS percentage
FROM orders;"""
a18=pd.read_sql(q18,conn)
a18
```

```python
# -- Find out the month in which the most orders were placed.
q19 = """SELECT MONTH(order_date) AS month, COUNT(*)
FROM orders 
GROUP BY MONTH(order_date)
LIMIT 1;"""
a19=pd.read_sql(q19,conn)
a19
```
